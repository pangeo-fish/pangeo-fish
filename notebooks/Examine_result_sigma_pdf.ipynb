## Collect configuration informations

```python
# Define the S3 path pattern for the configuration JSON files
config_path_pattern = "s3://gfts-ifremer/tags/bargip/*/configuration.json"

# List all configuration.json files in the specified S3 path
fs = fsspec.filesystem('s3')
config_files = fs.glob(config_path_pattern)

# Initialize a list to hold the data from each JSON file
data_list = []

# Read and parse each JSON file
for config_file in config_files:
    with fs.open(config_file, 'r') as f:
        config_data = json.load(f)
        # Extract the last part of the scratch_root path
        parsed_url = urlparse(config_data['scratch_root'])
        experiment = parsed_url.path.split('/')[-1]
        config_data['experiment'] = experiment
        data_list.append(config_data)

# Create a DataFrame from the list of JSON data
configurations = pd.DataFrame(data_list)

# Display the DataFrame
#df.head()  # Display the first few rows of the DataFrame
configurations.set_index('experiment', inplace=True)
```

## Get sigma, and check if it converged or not. 


```python
# Define the S3 path pattern to include all subdirectories
s3_base_path = 's3://gfts-ifremer/tags/bargip/*/*/parameters.json'

#s3_base_path = 's3://gfts-ifremer/tags/bargip/*/tracks_4/parameters.json'

# Create a filesystem object for S3
fs = fsspec.filesystem('s3')

# List all JSON files matching the pattern
files = fs.glob(s3_base_path)

# Initialize an empty list to store the data
data = []

# Read each JSON file and extract the value of "sigma"
for file in files:
    # Extract parts of the file path
    path_parts = file.split('/')
    experiment = path_parts[-3]
    tag_number = path_parts[-2]
    with fs.open(file, 'r') as f:
        content = f.read()
        data_json = json.loads(content)
        for key in data_json:
            sigma_value = data_json[key]['sigma']
            # Check if the sigma_value is almost the same as the max_sigma
            #
            max_sigma = 47.119351 #configurations.loc[experiment, 'max_sigma']
            if abs(sigma_value - max_sigma) <= 0.1:
                sigma_value = 'max_sigma'
    # Append the data to the list
    data.append([experiment, tag_number, sigma_value])

# Create a pandas DataFrame
df = pd.DataFrame(data, columns=['experiment', 'tag_number', 'sigma'])

# Pivot the DataFrame to get the desired format
df=df.pivot_table(index='tag_number', columns='experiment', values='sigma', aggfunc='first')

# Set tag_number as the index
df.reset_index(inplace=False)
df
#tag_table = df
```

## Add each simulation's information

### Compute pdf at the edge

```python
def compute_edge_pdf(tag_table, experiment ):
    # Initialize lists to store time information
    edge_sums = []
    # Find the first non-NaN experiment for each tag_number and extract time information
    # Iterate through the tag_number column
    tag_number=tag_table[experiment].first_valid_index()
    if experiment== 'tracks':
        bbox={'latitude': [42, 53], 'longitude': [-8, 4]}
    else: #tracks_2,_3, DK_test  , DK_test_2 
        bbox={'latitude': [40, 56], 'longitude': [-13, 5]}
    print(experiment,bbox)
    ds=(xr.open_dataset(
    f"s3://gfts-ifremer/tags/bargip/{experiment}/{tag_number}/emission.zarr",
                                    engine="zarr",
                                    chunks={},
                                    inline_array=True,
                                    storage_options=storage_options
                                )
    )
    # Define the tolerance
    tolerance = 0.025*2
    
    # Create a boolean DataArray for the 'edge' condition
    edge_condition = (
        ((ds.latitude >= bbox['latitude'][0] ) & (ds.latitude <= bbox['latitude'][0] + tolerance) |
         (ds.latitude >= bbox['latitude'][1] - tolerance) & (ds.latitude <= bbox['latitude'][1] )) |
        ((ds.longitude >= bbox['longitude'][0] ) & (ds.longitude <= bbox['longitude'][0] + tolerance) |
         (ds.longitude >= bbox['longitude'][1] - tolerance) & (ds.longitude <=  bbox['longitude'][1] ))
    )
    edge = xr.DataArray(edge_condition, dims=['y', 'x'], coords={'y': ds.y, 'x': ds.x}).compute()

    for tag_number in tag_table.index:
        print(experiment,tag_number)
        if not pd.isna(tag_table.loc[tag_number][experiment]) :#and tag_number not in excluded_tags:
            try:
                states=xr.open_dataset(
                        f"s3://gfts-ifremer/tags/bargip/{experiment}/{tag_number}/states.zarr",
                        engine="zarr",
                        chunks={},
                        inline_array=True,
                        storage_options=storage_options
                            ).states.isel(time=slice(0,-1))
                edge_sum=(states*(edge)).sum().compute().item()
                #non_edge_sum=(states*(~edge)).sum().compute().item()
                edge_average=edge_sum/states.time.size
                print(edge_average)                
                edge_sums.append(edge_average)                
            except Exception as e:
                print(f"Error processing {experiment}/{tag_number}: {e}")
                edge_sums.append(None)
        else:
            print(f"Error processing Nan at {experiment}/{tag_number}:")
            edge_sums.append(None)
    # Add time information to the DataFrame
    return edge_sums
```

```python
%%time
for experiment in ['tracks_2','tracks_3','tracks','DK_test','DK_test_2']:
    print(experiment+'_edge_array')
    tag_table[experiment+'_edge_sum'] = compute_edge_pdf(tag_table, experiment)
    orient='split'
    s3_path = 's3://gfts-ifremer/tags/bargip/summary_result.json'
    tag_table.to_json(s3_path, orient=orient, storage_options=storage_options)
```

### Add time of simulations.

```python
# Initialize lists to store time information
time_min_list = []
time_max_list = []
time_step_list = []
# Find the first non-NaN experiment for each tag_number and extract time information
# Iterate through the tag_number column
for tag_number in tag_table.index:
    first_experiment = tag_table.loc[tag_number].first_valid_index()
    #print(tag_number)
    if first_experiment:
        try:
            time = xr.open_dataset(
                f"s3://gfts-ifremer/tags/bargip/{first_experiment}/{tag_number}/states.zarr",
                engine="zarr",
                chunks={},
                inline_array=True,
                storage_options=storage_options
            ).time

            time_min = pd.Timestamp(time[0].values).strftime('%Y-%m-%dT%H')
            time_max = pd.Timestamp(time[-1].values).strftime('%Y-%m-%dT%H')
            time_step = time.size
            #print(first_experiment,tag_number,time.size,time_min,time_max)
            
            time_min_list.append(time_min)
            time_max_list.append(time_max)
            time_step_list.append(time_step)
        except Exception as e:
            print(f"Error processing {first_experiment}/{tag_number}: {e}")
            time_min_list.append(None)
            time_max_list.append(None)
            time_step_list.append(None)
    else:
        time_min_list.append(None)
        time_max_list.append(None)
        time_step_list.append(None)
# Add time information to the DataFrame
tag_table['time_min'] = time_min_list
tag_table['time_max'] = time_max_list
tag_table['time_step'] = time_step_list    
```
